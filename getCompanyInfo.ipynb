```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

url = 'https://dart.fss.or.kr/dsae001/main.do'
driver = webdriver.Chrome()
driver.get(url)
```

```python
from Investment import Analyzer
mdb = Analyzer.MarketDB()
companies = mdb.comp_codes()        # list

data = {'company_name': [], 'industry_text': []}
df = pd.DataFrame(data)
```

```python
i = 1
for company_name in companies:
    input_elements = driver.find_elements(By.CSS_SELECTOR, 'input.w40.p10')[0]
    input_elements.clear()
    input_elements.send_keys(f'{company_name}')
    button_elemnets = driver.find_elements(By.CLASS_NAME, 'btnSearch02')[0]
    button_elemnets.click()
    link_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//a[contains(text(), '동화약품')]"))
    )

    # 링크 클릭
    link_element.click()
    
    table_present = EC.presence_of_element_located((By.ID, 'corpDetailTabel'))
    WebDriverWait(driver, 10).until(table_present)
    
    html_content = driver.page_source
    print(html_content)

    #df = df.append({'company_name': company_name, 'industry_text': info_text}, ignore_index=True)
    #i += 1
    #print(f'[{i:04d}/{len(companies)} saving...]', end='\r')
    break
```
